```python
a = { i for i in range(10)}
b = {i for i in range(5)}
print(a)
print(b)
print("Intersection is " , a.intersection(b))
print("Union is " , a.union(b))
print("a diff b is " , a.difference(b))
print("symmetric difference a b is " , a.symmetric_difference(b))

a.difference_update(b)


flag1 = a.issubset(b)
flag2 = a.issuperset(b)
flag3 = a.isdisjoint(b)

print(flag1, flag2, flag3)
```

```python
# Converting list to  sets

a = [i*i for i in range(10)]
b = set(a)
print(b)

a.extend(a)
print(a)

b1 = set(a)
print(b1)
print("b1 is same as b. This means set has only unique values, i.e. keys")

```

```python
# checking conversion of sets back to lists

s1 = {0, 1, 64, 4, 36, 9, 16, 49, 81, 25}
s2 = [s1]
print(type(s2))
print(s2)
print("we created a list having set as an element")

# Retrying

s1 = {0, 1, 64, 4, 36, 9, 16, 49, 81, 25}
#format(s1, list)
#print(s1)  #wrong format function used. "Format" function is used only for conversion to strings

s1 = {0, 1, 64, 4, 36, 9, 16, 49, 81, 25}
s2 = list(s1)
print(s2)
print("this time, the set was converted to list. Hence the direct conversion of set to list is possible")
```

```python
# Find sum of all unique numbers in the list

a = [1,2,3,4,5,5,1,3,4,5,1,6]

sum1 =0
sum2 = 0
sum3 = 0

for i in a:
    sum1 +=i
print("sum of all elements is " , sum1)

# Method1: direct conversion of list to set

b = set(a)
for i in b:
    sum2 += i
print("sum of unique elemets by method 1 is " , sum2)

# Method 2:  creating a new set

b1 = set()

for i in a:
    b1.add(i)
for i in b1:
    sum3 += i
print("sum of unique elemets by Method 2 is " , sum3)
```

```python
# Finding no. of elements removed from array to make it a set

a = [int(i) for i in input().split()]

count1 = 0
count2 = 0

for i in a:
    count1 += 1
print("total list elements are " , count1)

b = set(a)
for i in b:
    count2 += 1

print("Total unique elements are " , count2)
print("elements removed are  " ,abs(count2 - count1))
```

```python
# Creating a set in same order of list elements

l = [1,4,2,1,4,5,6,7,0,9,4,9,3,0,2]
print(l)
s = set(l)
print(s)
print("Set assumes its own order (basically in ascending numeric order)")

# conversion to list order

s1 = set()
for i in l:
    s1.add(i)
print(s1)
print("again, set takes its own order")

l1 = []
for i in l:
    if i not in l1:
        l1.append(i)
print(l1)

s3 = set(l1)
print(s3)
print("may be, set will arrange its elements always, before forming")
```
