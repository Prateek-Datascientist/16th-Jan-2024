Given the "cars.csv" dataset, you need to analyze the car mileage in the city and on the highway. For this, you need to filter the dataset so that only car maker, city mileage and highway mileage are present. After this, you need to analyze the duplicate values present in this filtered dataset.
Print the number of duplicated rows in the filtered dataset.

```python
df1 = df.filter(items = ['make', 'city-mpg' ,'highway-mpg' ])
a = df1[df1.duplicated() == True]
a.make.value_counts()
```
