```python
## Open and read data file as specified in the question
## Print the required output in given format

import pandas as pd
cols = [ 'SL' , 'SW' , 'PL' , 'PW' , 'variety']
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names = cols)
file = data.copy()

file1 = file[ file.variety == "Iris-setosa"]
file2 = file[ file.variety == "Iris-versicolor"]
file3 = file[ file.variety == "Iris-virginica"]

file1 = file1.describe().loc[["min" , "max" , "mean"]].values
file2 = file2.describe().loc[["min" , "max" , "mean"]].values
file3 = file3.describe().loc[["min" , "max" , "mean"]].values

for i in file1:
    for j in i:
        print(j, end= ' ')
        
    print("Iris-setosa")
    
for i in file2:
    for j in i:
        print(j, end= ' ')
         
    print("Iris-versicolor")

for i in file3:
    for j in i:
        print(j, end= ' ')
        
    print("Iris-virginica")
```

```python
import pandas as pd
import numpy as np

a1 = pd.DataFrame( { 'A' :[1,2,3] , 'B': [4,5,6] , 'C': [7,8,9]  } , index = ['a' ,'b' ,'c'])

def f1(n):
    a1 = (n+1)**2
    if a1%5 == 0:
        return "divisible by 5"
    else: return "not divisible"

a1["B+1 whole square"] = a1.B.apply(f1)   #B.apply gives us a new column

a1
```
