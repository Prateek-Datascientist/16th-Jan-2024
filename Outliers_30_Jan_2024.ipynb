Problem statement
You have been provided with an array of data points. Your task is to analyze this array and find out the outliers present. To analyze the outliers you need to use the upper threshold and lower threshold values of 3 and -3 respectively.

data = [1, 2, 2, 2, 3, 1, 1, -19, 2, 2, 2, 3, 1, 1, 2, 20, 4, 2, 3, -3, 5, -2, -1, -2]
Expected Output
Print the outliers separated by lines.
Also, plot the distribution curve for the given array.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 2, 3, 1, 1, -19, 2, 2, 2, 3, 1, 1, 2, 20, 4, 2, 3, -3, 5, -2, -1, -2]
```

```python
std_dev = np.std(data)
mean1 = np.mean(data)
for i in data:
  if abs((i-mean1)/std_dev) > 3:
    print(i)
sns.distplot(data)
```

# Replacing outliers of Item_Outlet_Sales by max(value) of remaining data set

```python
z_score =  stats.zscore(df.Item_Outlet_Sales)
df1 = df.copy()
import matplotlib.pyplot as plt

outliers_index = z_score[ abs(z_score) > 3 ].index

non_out = df.Item_Outlet_Sales[ abs(z_score) < 3 ]

max_val  = non_out.max()
print("New maximum value is   " ,max_val)
print("Old maximum value was " , df.Item_Outlet_Sales.max())
df1.loc[outliers_index , "Item_Outlet_Sales" ] = max_val
df1
sns.boxplot(x =df1.Item_Outlet_Sales)
plt.show()
sns.boxplot(x =df.Item_Outlet_Sales)
plt.show()
df1
```
