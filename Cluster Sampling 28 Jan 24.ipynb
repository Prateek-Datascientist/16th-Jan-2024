```python
import pandas as pd
import numpy as np
```

```python

df = pd.DataFrame({'tour': np.repeat(np.arange(1,11), 30),
                   'experience': np.random.normal(loc=7, scale=1, size=300)})

# We use np.random to choose from an array(numpy). But we use x.sample(2) to choose from a pandas data frame
clusters = np.random.choice(np.arange(1,11) , size = 4)

cluster_sample = df[df['tour'].isin(clusters)]

cluster_sample['tour'].value_counts()
```
