```python
x = [1,2,3,4]
y = [55,33,22,55]
plt.plot(x,y, 'b-.' , marker = 'p' , label = "Line1"  )
plt.legend()
plt.title("Random Graph")
#plt.legend()    # The legend function helps to showcase label / legend on graph
plt.show()
```

```python
a1 = np.arange(0, 5, 0.5)
b1 = a1**3
b2 = a1**2
plt.text(0,30, "Welcome Viewer")
plt.plot(a1,b1, 'r--', label = 'x^3'  )
plt.plot(a1, b2,  'y', label = 'x^2' )
plt.legend()
plt.grid( axis = 'x')
plt.scatter(a1, b1 , s = b1)
plt.scatter(a1, b2 , s = b2)
for i in range(len(a1)):
    plt.text(a1[i]+0.1 , b1[i]+0.1 , int(b1[i]) )
for i in range(len(a1)):
    plt.text(a1[i]+0.1 , b2[i]+0.1 , int(b2[i]) )
plt.title(" x^3   v/s   x^2")
plt.show()
```

```python
# Bubble plot

x = np.arange(0, 10 , 1)
y = x
z = x
col = [np.arange(0, len(x))]
plt.scatter( x,y , c = col, marker = 'p', alpha = 1 , edgecolor = 'r', linewidth = 0.5)
for i in range(len(x)):
    plt.text(x[i]+ 0.1, y[i]+0.1, z[i])
plt.xlabel("x axis")
plt.ylabel("y axis")
```
