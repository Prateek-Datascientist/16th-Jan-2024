## First Machine learning Model (Linear Regression)

```python
d1 =  datasets.load_breast_cancer()
data = d1.data
target = d1.target
print(type(d1))
print(type(data))

print(data.shape)
print(target.shape)

df = pd.DataFrame(data, columns= d1.feature_names)
print(df.shape)
from sklearn.linear_model import  LinearRegression
alg1 = LinearRegression()

from sklearn.model_selection import train_test_split
print(type(alg1))
print(alg1)

a = train_test_split(data , target)
```

```python
print(type(a))
xtrain, xtest, ytrain, ytest = a
print(type(xtrain))
print(type(ytrain))
print(xtrain.shape)
print(xtest.shape)
print(data.shape)
print(type(xtrain.shape))
```

```python
alg1.fit(xtrain, ytrain)
```

```python
ypred = alg1.predict(xtest)
```

```python
print(type(ypred))
print(ypred.shape)
# Comparing y test and y predict
```

```python
import seaborn as sns
sns.scatterplot(x=ytest, y=ypred)
```

```python
mean1 = ytest.mean()
mean2 = ypred.mean()
print(mean1)
print(mean2)
print(int((abs(mean1 - mean2))*100/mean1), end= '%')
```

## Model 2 of ML with greater xtrain percentage (linear regression)

```python
b = train_test_split(data , target, train_size= 510)
```

```python
xtrain, xtest, ytrain, ytest = b
alg1.fit(xtrain, ytrain)
ypred2 = alg1.predict(xtest)
mean3 = ypred2.mean()
print(mean1)
print(mean2)
print(mean3)
print(int(abs(mean3 - mean1)*100/mean1) , end= '%  We get 1% greater accuracy')
```
